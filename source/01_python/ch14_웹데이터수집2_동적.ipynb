## 3.2 예제

### 1) 다음뉴스 검색

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.get('https://www.daum.net')
driver.find_element(By.NAME, "q").click
quary = input('검색할 단어 : ')
driver.find_element(By.CSS_SELECTOR, 'input[type=text]').clear()
driver.find_element(By.CSS_SELECTOR, 'input[type=text]').send_keys(quary)
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
time.sleep(2)
driver.find_element(By.LINK_TEXT, '뉴스').click()
time.sleep(2)
news_list = []
pages = 3
for page in range(1, pages+1):
    elems = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong.tit-g > a')
    for elem in elems:
        title = elem.text
        link = elem.get_attribute('href')
        news_list.append([title,link])
        print(title,link)
    page_div = driver.find_element(By.CSS_SELECTOR, 'div.inner_paging')
    next_page = page_div.find_element(By.LINK_TEXT,str(page+1))
    next_page.click()
    time.sleep(1.5)
````

````python
import pandas as pd
pd.DataFrame(news_list, columns=['제목','링크'])
````

### 2) 맞춤법검사기

````python
# ch_맞춤법 전 300자 이내로 맞추기
def text_spliter(path):
    with open(path, 'r', encoding='utf-8') as f:
        text = f.read()
    text_list = [] # 맞춤법 감사할 text 내용(300이자 이내로 list)
    while len(text) > 300 :
        temp = text[:300]
        new_line_char_index = temp.rfind('.')+1
        text_list.append(text[:new_line_char_index])
        text= text[new_line_char_index:]
    text_list.append(text)
    return text_list
````

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get('https://www.naver.com')
driver.find_element(By.NAME, "query").click
driver.find_element(By.CSS_SELECTOR, 'input[type=search]').send_keys('맞춤법검사기')
driver.find_element(By.CLASS_NAME, 'btn_search').click()
time.sleep(2)
result = ''
for t in text_spliter('data/ch14/맞춤법전.txt'):
    driver.find_element(By.CSS_SELECTOR, 'textarea').clear()
    driver.find_element(By.CSS_SELECTOR, 'textarea').send_keys(t)
    driver.find_element(By.CLASS_NAME, 'btn_check').click()
    time.sleep(1)
    result += (driver.find_element(By.CLASS_NAME, '_result_text').text)

driver.quit()
with open('data/ch14/맞춤법후.txt', 'w') as f:
    f.write(result)
````

### 3) 번역기

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get('https://www.google.com/')
time.sleep(2)
driver.find_element(By.CSS_SELECTOR, "textarea").send_keys('번역기')
time.sleep(2)
driver.find_element(By.CSS_SELECTOR, "textarea").send_keys(Keys.RETURN)
time.sleep(2)
````

````python
result = ''
driver.find_element(By.CSS_SELECTOR, 'span.tw-menu-btn-image > svg').click()
for t in text_spliter('data/ch14/맞춤법전.txt'):
    driver.find_element(By.CSS_SELECTOR, 'textarea.tw-ta').clear()
    driver.find_element(By.CSS_SELECTOR, 'textarea.tw-ta').send_keys(t)
    time.sleep(2)
    result += (driver.find_element(By.CSS_SELECTOR, 'pre.tw-data-text > span').text)

driver.quit()
with open('data/ch14/번역후.txt', 'w') as f:
    f.write(result)
````
