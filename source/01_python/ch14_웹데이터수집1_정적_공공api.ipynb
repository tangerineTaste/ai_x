````python
# 다음 뉴스 검색(키워드, 원하는 페이지수)
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
def collect_list(keyword, page=1):
    'keyword로 다음 검색한 결과를 return'
    url = f'https://search.daum.net/search?nil_suggest=btn&w=news&DA=PGD&sq=elden&o=1&sugo=11&cluster=y&q={keyword}&p={page}'
    params = {'q':keyword, 'p':page}
    response = requests.get(url,params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    element = soup.select('div.item-title > strong.tit-g.clamp-g > a')
    item_find_list = []
    for idx, item in enumerate(element):
        item_find_list.append({'no':(page-1)*10+idx,'title':item.text,'link':item.attrs.get('href')})
    return item_find_list
````

````python
# keyword를 display수만큼 지식in 검색 결과를 데이터프레임에 (방법1 크롤링)
import requests
import pandas as pd
import json
from dotenv import load_dotenv
import os
def naver_in_get(keyword, display): 
    load_dotenv()
    client_id = os.getenv('Client_ID')
    client_secret = os.getenv('Client_Secret')
    url = f'https://openapi.naver.com/v1/search/kin.json?query={keyword}&display={display}'
    headers = {'X-Naver-Client-Id' : client_id, "X-Naver-Client-Secret" : client_secret}
    response = requests.get(url, headers=headers)
    items = response.json()['items']
    items_list =[]
    for item in items:
        items_list.append([item.get('title').replace('<b>','').replace('</b>',''),
                            item.get('link'),
                            item.get('description').replace('<b>','').replace('</b>','')])
    return items_list
        
pd.DataFrame(naver_in_get('챗지피티',30), columns=['제목','링크','내용']).sample(5)

````

### quiz) 네이버 open API를 이용해서 청바지 이미지 100건 데이터를 csv파일로"
    데이터 : 제목, 이미지링크, 썸네일링크, 높이, 폭

````python
import requests
def save_img(attr, idx, query, url):
    '{attr}_{idx}_{query}.{확장자} 이미지 저장'
    file_extension = url.split('.')[-1]
    quote_index = file_extension.find('?')
    
    if quote_index != -1:
        file_extension = file_extension[:quote_index]
    img = requests.get(url).content
    open(f'./data/ch14/청바지/{attr}_{idx}_{query}.{file_extension}', 'wb').write(img)
````

````python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
def jean(keyword, display):
    load_dotenv()
    client_id = os.getenv('Client_ID')
    client_secret = os.getenv('Client_Secret')
    url = f'https://openapi.naver.com/v1/search/image?query={keyword}&display={display}'
    headers = {'X-Naver-Client-Id' : client_id, "X-Naver-Client-Secret" : client_secret}
    response = requests.get(url, headers=headers)
    items = response.json()['items']
    items_list =[]
    for idx,item in enumerate(items):
        items_list.append([item.get('title').replace('<b>','').replace('</b>',''),
                            item.get('link'),
                            item.get('thumbnail'),
                            item.get('sizeheight'),
                            item.get('sizewidth')
                            ])
        save_img('메인',idx+1,keyword,item.get('link'))
        save_img('메인',idx+1,keyword,item.get('thumbnail'))
        if idx/20 == round(idx/20):
            print(f'{idx}% 진행 중 입니다')
    pd.DataFrame(items_list, columns=['제목','링크','썸네일','높이','폭']).to_csv('data/ch14/청바지이미지.csv',index=False,encoding='utf-8')
    print('csv파일 저장 완료!')
````

````python
import pandas as pd
url2 = f'http://ws.bus.go.kr/api/rest/busRouteInfo/getStaionByRoute?ServiceKey={key}&busRouteId={busRouteId}'
response = requests.get(url2)
soup = BeautifulSoup(response.text, 'xml')
bus_station = []
for item in soup.select('itemList'):
    stationNm = item.select_one('stationNm').text
    station = item.select_one('station').text
    gpsX = item.select_one('posX').text
    gpsY = item.select_one('posY').text
    bus_station.append([stationNm,station,gpsX,gpsY])
df_station = pd.DataFrame(bus_station,columns=['정류소명','ID','경도','위도']) 
````

````python
#step3 : busRouteId로 버스 위치 정보 조회 
url3 = f'http://ws.bus.go.kr/api/rest/buspos/getBusPosByRtid?ServiceKey={key}&busRouteId={busRouteId}'
response = requests.get(url3)
soup = BeautifulSoup(response.text, 'xml')
bus_position = []
for item_list in soup.select('itemList'):
    plain_no = item_list.select_one('plainNo').text
    congetion = item_list.select_one('congetion').text
    congetion = '여유' if congetion =='3' else '보통' if congetion=='4' else '혼잡' if congetion=='5'else '매우혼잡'
    posX = item_list.select_one('posX').text
    posY = item_list.select_one('posY').text
    
    lastStnId = item_list.select_one('lastStnId').text
    lastStnId = df_station.loc[df_station['ID'] == lastStnId,'정류소명'].iloc[0]

    nextStId = item_list.select_one('nextStId').text
    nextStId = df_station.loc[df_station['ID'] == nextStId,'정류소명'].iloc[0]
    
    nextStTm = str(int(item_list.select_one('nextStTm').text)//60) + '분'
    bus_position.append([plain_no,congetion,posX,posY,lastStnId,nextStId,nextStTm])
df_pos = pd.DataFrame(bus_position, columns=['차량번호','혼잡도','경도','위도','이전정류장','다음정류장','도착소요시간'])
df_pos.sample(5)
````

# 4절 연습문제
- YES24 베스트셀러 정보를 ch14_yes24_.csv로 출력
- 순위, 책이름, 저자, 출판사, 가격 (1~48위) => 리스트 => 데이터프레임 => csv

````python
import requests
import pandas as pd
from bs4 import BeautifulSoup

element = []
result = []
pages = 2
for page in range(1, pages+1):
    url = f'https://www.yes24.com/product/category/bestseller?categoryNumber=001&pageNumber={page}&pageSize=24'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    element.extend(soup.select('div.item_info'))
for idx,e in enumerate(element):
    title = e.select_one('a.gd_name').text
    auth = e.select_one('div.info_row > span.info_auth').text.strip()
    pub = e.select_one('div.info_row > span.info_pub').text.strip()
    price = e.select_one('div.info_row > strong.txt_num ').text
    result.append([idx+1,title, auth, pub, price])
pd.DataFrame(result, columns=['순위','책이름','저자','출판사','가격']).to_csv('data/ch14/예스24_책순위.csv',index=False)
pd.DataFrame(result, columns=['순위','책이름','저자','출판사','가격'])
````
