````python
# soup.select('선택자') : 해당 선택자 모든 엘리먼트를 리스트로
el = soup.select('h1')
print('리스트 el : ' , el)
print('el의 text들 : ', [e.text for e in el ])
print('el의 속성들 : ', [e.attrs for e in el])
print('el의 class 속성들 : ', [e.attrs.get('class') for e in el ])
````

````python
# select_one(선택자) 와 find(태그, 속성)
print('select_one : ',soup.select_one('h1.css'))
print('find : ', soup.find('h1',{'class':'css'}))
print('find :', soup.find('h1', class_ = 'css'))
print()
print('select_one : ', soup.select_one('h1#text'))
print('find : ',soup.find('h1', {'id':'text'}))
print()
print('select')
````

````python
# select(선택)와 find_all(태그, 속성)
print('모든 h1.css, span 태그(select) : \n',soup.select('h1.css, span'))
print()
print('모든 h1.css, span태그(find_all) : \n',soup.find_all(['h1','span'],class_='css'))
print()
````

````python
# 없는 엘리먼트 찾기
print('find_all(빈list) : ', soup.find_all('a', class_='css'))
print('find(None) : ', soup.find('a', class_ = 'css'))
print('select(빈list) : ', soup.select('a'))
print('select(None) :')
````

### 2) 이번주 로또번호 출력
- google에 "로또번호 당첨" 검색
```
1173회 당첨결과 (2025년 05월 24일 추첨)
당첨번호[1,5,18,20,30,35]
보너스 3
```

````python
# 방법1 (select)
import requests
from bs4 import BeautifulSoup
url = 'https://dhlottery.co.kr/gameResult.do?method=byWin&wiselog=H_C_1_1'
response = requests.get(url)
soup = BeautifulSoup(response.text,
                    'html.parser')              
count = soup.select_one('div.win_result > h4').text
date = soup.select_one('div.win_result > p').text
balls = [int(ball.text) for ball in soup.select('div.num> p > span')]
win_balls = balls[:-1]
bonus_ball = balls[-1] 
print('{} {}\n당첨번호 {}\n보너스 {}'.format(count,date,win_balls,bonus_ball))
````

````python
# 다음 뉴스 검색(키워드, 원하는 페이지수)
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
def collect_list(keyword, page=1):
    'keyword로 다음 검색한 결과를 return'
    url = f'https://search.daum.net/search?nil_suggest=btn&w=news&DA=PGD&sq=elden&o=1&sugo=11&cluster=y&q={keyword}&p={page}'
    params = {'q':keyword, 'p':page}
    response = requests.get(url,params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    element = soup.select('div.item-title > strong.tit-g.clamp-g > a')
    item_find_list = []
    for idx, item in enumerate(element):
        item_find_list.append({'no':(page-1)*10+idx,'title':item.text,'link':item.attrs.get('href')})
    return item_find_list
````

````python
# 원하는 keyword로 다음검색 
result = []
keyword = '밤에 통치자'
pages = 3
for page in range(1, pages+1):
    result.extend(collect_list(keyword, page))
    time.sleep(1)
pd.DataFrame(result)
````
